```python
import matplotlib.pyplot as plt
import numpy as np

# --- Datos ---
periodo = np.arange(1, 25)
valor = np.array([
    10.2, 9.15, 15.5, 10.15, 14.5, 17.0, 9.0, 10.0, 12.0, 13.0,
    14.0, 20.0, 21.0, 26.0, 25.0, 30.5, 25.5, 26.0, 27.5, 22.0,
    26.5, 30.5, 26.5, 33.0
])

# --- Crear la figura automáticamente ---
fig, ax = plt.subplots(figsize=(8, 4))

# --- Graficar directamente con estilo automático ---
ax.plot(periodo, valor, 'o-', color='blue', mfc='blue', mec='blue')

# --- Personalizar usando funciones automáticas ---
ax.set(
    title='Líneas: serie temporal con tendencia y estacionalidad',
    xlabel='Periodo',
    ylabel='Valor'
)

# Deja que Matplotlib ajuste los límites, pero añade una ligera expansión
ax.margins(x=0.05, y=0.1)

# Activa la cuadrícula ligera
ax.grid(True, linestyle='--', alpha=0.5)



# Mostrar
plt.show()
```
